```python
import os
import glob
def get_file_list(path: str, wild: str = None) -> list:
    ret_list = []
    if not path.endswith('/'):
        path = path + '/'
    
    if wild:
        for filename in glob.glob(path + wild):
            ret_list.append(filename)
    else:
        for filename in os.listdir(path):
            ret_list.append(filename)
    
    return ret_list
```

```python
import pandas as pd
import glob
def filter_dicom(path=None, glob_str=None, element=None) -> pd.DataFrame:
    if not path:
        path = input("Enter full image directory path: ")
    if not glob_str:
        glob_str = input("Enter glob string: ")
    if not element:
        element = input("Enter data element to print: ")
    
    elems = []
    filenames = []
    for file in get_file_list(path, glob_str):
        ds = pydicom.read_file(file, force=True)
        elem = []
        
        if element in ds:
            #elem = ds[element].value
            elem.append(ds[element].value)
        else:
            #elem = "NULL"
            elem.append("NULL")
            
        base = os.path.basename(file)
        filenames.append(base)
        elem.append(base[0:2])
        elems.append(elem)

    return pd.DataFrame(elems, filenames, columns=[element, 'Site'])
    
```
